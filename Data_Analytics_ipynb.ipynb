# Etapa 2

## Limpieza de Datos

```python
# Eliminación de duplicados
clientes_limpio = clientes.drop_duplicates()
ventas_limpio = ventas.drop_duplicates()
marketing_limpio = marketing.drop_duplicates()

print("Duplicados eliminados:")
print("Clientes:", clientes.shape[0] - clientes_limpio.shape[0])
print("Ventas:", ventas.shape[0] - ventas_limpio.shape[0])
print("Marketing:", marketing.shape[0] - marketing_limpio.shape[0])
```

```python
# Limpieza de caracteres no deseados
# Por ejemplo, eliminar espacios en blanco, símbolos o caracteres especiales en columnas de texto
for df in [clientes_limpio, ventas_limpio, marketing_limpio]:
    df.columns = df.columns.str.strip()  # limpia espacios en nombres de columnas
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].astype(str).str.replace('[^a-zA-Z0-9áéíóúÁÉÍÓÚñÑ\s\.,-]', '', regex=True).str.strip()

print("✔ Limpieza de texto y columnas realizada.")

# Documentación del proceso
documentacion_limpieza = pd.DataFrame({
    'Dataset': ['Clientes', 'Ventas', 'Marketing'],
    'Duplicados Eliminados': [
        clientes.shape[0] - clientes_limpio.shape[0],
        ventas.shape[0] - ventas_limpio.shape[0],
        marketing.shape[0] - marketing_limpio.shape[0]
    ],
    'Filas Finales': [
        clientes_limpio.shape[0],
        ventas_limpio.shape[0],
        marketing_limpio.shape[0]
    ]
})

display(documentacion_limpieza)
```

## Transformación de Datos

```python
# The error "KeyError: 'total_venta'" occurs because the column 'total_venta'
# does not exist in the 'ventas_limpio' DataFrame at the time it's being accessed.
# This usually happens when the code that's supposed to create this column
# has not been executed yet.

# To fix this, we will first ensure 'ventas_limpio' is properly prepared
# for the 'total_venta' calculation within this cell.
# This involves handling any remaining missing values in 'precio' and 'cantidad'
# and then converting 'precio' to a numeric type if it's not already.

# It's good practice to work on a copy to avoid SettingWithCopyWarning.
ventas_limpio_procesado = ventas_limpio.copy()

# Handle missing values in 'precio' and 'cantidad' before calculation.
ventas_limpio_procesado = ventas_limpio_procesado.dropna(subset=['precio', 'cantidad'])

# Convert 'precio' to float. It was initially an object (string) and
# previous cleaning steps removed non-numeric characters like '$'.
ventas_limpio_procesado['precio'] = ventas_limpio_procesado['precio'].astype(float)

# Calculate 'total_venta'
ventas_limpio_procesado['total_venta'] = ventas_limpio_procesado['precio'] * ventas_limpio_procesado['cantidad']

# Now perform the original calculations using the prepared DataFrame
# Calcular las ventas promedio generales
promedio_ventas = ventas_limpio_procesado['total_venta'].mean()

# Filtrar productos con un rendimiento superior al promedio
productos_alto_rendimiento = ventas_limpio_procesado[ventas_limpio_procesado['total_venta'] > promedio_ventas]

print(f"Promedio general de ventas: ${promedio_ventas:,.2f}")
print(f"Productos de alto rendimiento: {productos_alto_rendimiento.shape[0]} filas")

display(productos_alto_rendimiento.head())

# Note: The original 'ventas_limpio' DataFrame outside this cell will not be modified.
# If you wish to update 'ventas_limpio' globally for subsequent cells,
# you would explicitly assign: ventas_limpio = ventas_limpio_procesado
```

```python
# Crear ventas_limpio gestionando valores faltantes y duplicados
ventas_limpio = ventas.dropna(subset=['precio', 'cantidad']).drop_duplicates().copy()

# Convierte 'precio' en numérico eliminando '$' y convirtiendo a flotante
ventas_limpio['precio'] = ventas_limpio['precio'].astype(str).str.replace('$', '', regex=False).astype(float)

# Crear una métrica de rendimiento (por ejemplo, ventas totales por producto)
ventas_limpio['total_venta'] = ventas_limpio['precio'] * ventas_limpio['cantidad']

print("Dimensiones de ventas_limpio después de la limpieza y transformación:", ventas_limpio.shape)
display(ventas_limpio.head())
```

## Agregación de Datos

```python
# Agregación por categoría
ventas_por_categoria = ventas_limpio.groupby('categoria').agg({
    'total_venta': ['sum', 'mean', 'count']
}).reset_index()

# Renombrar columnas para mayor claridad
ventas_por_categoria.columns = ['Categoria', 'Ingreso_Total', 'Promedio_Venta', 'Cantidad_Transacciones']

# Ordenar por mayor ingreso
ventas_por_categoria = ventas_por_categoria.sort_values(by='Ingreso_Total', ascending=False)

display(ventas_por_categoria)

print("✔ Resumen de ventas por categoría generado.")
```

## Integración de Datos

```python
# Unir los datasets (por ejemplo, por producto)
datos_integrados = pd.merge(ventas_limpio, marketing_limpio, on='producto', how='left')

# Visualizar el resultado
print("Dimensiones del dataset integrado:", datos_integrados.shape)
display(datos_integrados.head())

# Análisis básico post-integración
print("Columnas combinadas:", datos_integrados.columns.tolist())
print("Valores nulos después de la integración:")
print(datos_integrados.isnull().sum())
```

# Etapa 3

## Estadística Descriptiva

```python
# Estadística descriptiva sobre ventas (precio, cantidad, total_venta)
desc = ventas_limpio[['precio','cantidad','total_venta']].describe().T
# Medidas adicionales
from scipy import stats
import numpy as np

def iqr(x): return np.nanpercentile(x,75) - np.nanpercentile(x,25)

extra = []
for col in ['precio','cantidad','total_venta']:
    serie = ventas_limpio[col].dropna()
    extra.append({
        'variable': col,
        'median': float(serie.median()),
        'mode': serie.mode().iloc[0] if not serie.mode().empty else np.nan,
        'variance': float(serie.var()),
        'std': float(serie.std()),
        'iqr': float(iqr(serie)),
        'skew': float(serie.skew()),
        'kurtosis': float(serie.kurtosis())
    })
extra_df = pd.DataFrame(extra).set_index('variable')

print("Resumen estadístico (describe):")
display(desc)
print("\nMedidas adicionales:")
display(extra_df)
```

## Visualización Interactiva con Plotly

```python
# 1) Asegurar ventas_limpio y columnas de fecha/periodo
import pandas as pd
import plotly.express as px

# si ya lo tenés cargado, esto no rompe nada; si no, carga desde Drive
try:
    ventas_limpio
except NameError:
    ventas = pd.read_csv('/content/drive/MyDrive/datasets/ventas.csv')
    ventas_limpio = ventas.drop_duplicates().dropna(subset=['precio','cantidad']).copy()
    ventas_limpio['precio'] = ventas_limpio['precio'].astype(str).str.replace('$','', regex=False).astype(float)
    ventas_limpio['total_venta'] = ventas_limpio['precio'] * ventas_limpio['cantidad']

# normalizar fechas y crear periodos
ventas_limpio['fecha_venta'] = pd.to_datetime(ventas_limpio['fecha_venta'], dayfirst=True, errors='coerce')
ventas_limpio = ventas_limpio.dropna(subset=['fecha_venta'])  # quitar filas con fecha inválida si las hay
ventas_limpio['month'] = ventas_limpio['fecha_venta'].dt.to_period('M').dt.to_timestamp()
```

```python
# 2) Dashboard interactivo corregido
import ipywidgets as widgets
from IPython.display import display, clear_output
import plotly.express as px

# widget
categoria_options = ['Todos'] + sorted(ventas_limpio['categoria'].dropna().unique().tolist())
w_cat = widgets.Dropdown(options=categoria_options, description='Categoría:')

out = widgets.Output()

def actualizar_dashboard(cat):
    with out:
        clear_output(wait=True)
        df = ventas_limpio.copy()
        if cat != 'Todos':
            df = df[df['categoria'] == cat]
        # evitar error si df queda vacío
        if df.empty:
            print("No hay datos para la categoría seleccionada.")
            return

        mensual = df.groupby('month')['total_venta'].sum().reset_index()
        fig1 = px.line(mensual, x='month', y='total_venta', title=f'Ingresos mensuales ({cat})')
        top = df.groupby('producto')['total_venta'].sum().nlargest(10).reset_index()
        fig2 = px.bar(top, x='total_venta', y='producto', orientation='h', title=f'Top 10 productos ({cat})')
        fig3 = px.scatter(df, x='precio', y='cantidad', hover_data=['producto'], title=f'Precio vs Cantidad ({cat})')

        # usar show para forzar render en Colab
        fig1.show()
        fig2.show()
        fig3.show()

# observador explícito (names='value' es importante)
def on_change(change):
    if change.get('name') == 'value':
        actualizar_dashboard(change.get('new'))

w_cat.observe(on_change, names='value')

# mostrar widget + out
display(w_cat)
display(out)

# inicializar
actualizar_dashboard('Todos')
```
